5. Write a function ``find_borough(desc)`` that takes 1 *string* argument and returns one of the districts defined in the ``boroughs`` list. The function is to return the first (in order) name of the district, which is contained in ``desc``. If no name is found, return *Other*.

```python
def find_borough(desc):
    boroughs = ['Stare Miasto',
                 'Wilda',
                 'Jeżyce',
                 'Rataje',
                 'Piątkowo',
                 'Winogrady',
                 'Miłostowo',
                 'Dębiec']
    for item in boroughs:
        if item in desc:
            return item
    return 'Other'
```

6. Add the ``Borough`` column which will contain the information about the districts and will be derived from the ``Localization`` column. Use the ``find_borough`` function for this.

```python
df['Borough'] = df['Location'].apply(find_borough) 
df
```
